```python
import pandas as pd
cities = ['New York', 'London', 'Paris', 'Tokyo', 'Sydney']

#Create a series from the list
cities_series = pd.Series(cities)

#Outputting the series
cities_series
```

```python
# Create a dictionary of animals and their sounds
animal_sounds = {
    'Dog': 'Bark',
    'Cat': 'Meow',
    'Cow': 'Moo',
    'Lion': 'Roar',
    'Elephant': 'Trumpet'
}

# Create a pandas Series from the dictionary
animal_sounds_series = pd.Series(animal_sounds)

# Display the pandas Series
animal_sounds_series
```

```python
# Create a dictionary of days and dates starting from April 7th, 2001
days_dates = {
    'Monday': pd.Timestamp('2001-04-07'),
    'Tuesday': pd.Timestamp('2001-04-08'),
    'Wednesday': pd.Timestamp('2001-04-09'),
    'Thursday': pd.Timestamp('2001-04-10'),
    'Friday': pd.Timestamp('2001-04-11')
}

# Create a pandas Series from the dictionary
days_dates_series = pd.Series(days_dates)

# Display the pandas Series
days_dates_series
```

```python
# Create a dictionary of days and dates
days_dates = {
    'Monday': '2023-07-03',
    'Tuesday': '2023-07-04',
    'Wednesday': '2023-07-05',
    'Thursday': '2023-07-06',
    'Friday': '2023-07-07'
}

# Filter the dictionary for Monday and Tuesday
filtered_days_dates = {key: value for key, value in days_dates.items() if key in ['Monday', 'Tuesday']}

# Create a pandas Series from the filtered dictionary
filtered_days_dates_series = pd.Series(filtered_days_dates)

# Display the pandas Series
filtered_days_dates_series
```

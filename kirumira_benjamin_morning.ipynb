```python
#context manager for file handling that automatically opens and closes a file:
def open_file(filename):
    with open(filename, "r") as file:
        data = file.read()
    return data


def main():
    data = open_file("myfile.txt")
    print(data)


if __name__ == "__main__":
    main()

#Context manager for managing a database connection:
import sqlite3


def connect_to_database():
    connection = sqlite3.connect("mydatabase.db")
    with connection:
        cursor = connection.cursor()
        cursor.execute("CREATE TABLE IF NOT EXISTS users (id INTEGER PRIMARY KEY, name TEXT)")


def main():
    connect_to_database()


if __name__ == "__main__":
    main()


#Multithreading and multiprocessing that allows us to run the function for different amounts of time:
import threading
import time


def long_running_function(name):
    time.sleep(3)
    print(f"Thread {name} is finished.")


def main():
    threads = []
    for i in range(3):
        thread = threading.Thread(target=long_running_function, args=(i,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()


if __name__ == "__main__":
    main()
```
